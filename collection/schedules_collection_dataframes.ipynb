```python
import basketball_reference_web_scraper.readers as br
import pandas as pd
import ast 
import json
```

```python
twelve = br.return_json_encoded_schedule(2012)
```

```python
final_twelve = ast.literal_eval(twelve)['parsed_event_list']
with open('twelve.json', 'a') as _f:
    _f.write(json.dumps(final_twelve) + '\n')
```

```python
loaded_twelve = []
with open('twelve.json', 'r') as _f:
    loaded_schedule = json.load(_f)
    for game in loaded_schedule:
        loaded_twelve.append(ast.literal_eval(game))
```

```python
df_twelve = pd.DataFrame(loaded_twelve)
df_twelve['start_time'] = pd.to_datetime(df_twelve['start_time'])
df_twelve.head()
```

```python
twelve_unique_dates = df_twelve['start_time'].map(lambda t:t.date()).unique()
print(twelve_unique_dates)
```

```python
df_twelve.to_csv('2012_schedule.csv')
```

```python
pd.read_csv('2012_schedule.csv', index_col=0)
```
