```python
import pandas as pd
import numpy as np
import json
import ast
import basketball_reference_web_scraper.readers as br
```

```python
twelve = br.return_json_encoded_all_player_season_statistics(2012)
```

```python
final_twelve = []
for x in twelve:
    final_twelve.append(ast.literal_eval(x))
with open('twelve_player_stats.json', 'a') as _a:
    _a.write(json.dumps(final_twelve))
```

```python
df_twelve_players = pd.DataFrame(final_twelve)
df_twelve_players
```

```python
df_twelve_players.to_csv('2012_player_stats.csv')
```

```python
pd.read_csv('2012_player_stats.csv', index_col=0)
```
